```python
# Import lớp Image
from PIL import Image
```

```python
# Thư mục chứa hình ảnh
dir = 'C:/Pictures'
```

```python
# Đường dẫn đến hình ảnh
image_path = dir +'/1.PNG'
print(image_path)
```

```python
# Sử dụng Image.open() đọc ảnh
img01 = Image.open(image_path)
```

```python
# Hiển thị hình ảnh
display(img01)
```

```python
# In định dạng ảnh
print("Định dạng ảnh: ", img01.format)
```

```python
# In ra size
print("Kích thước hình ảnh: ", img01.size)
```

```python
# Đóng tập tin
img01.close()
```
